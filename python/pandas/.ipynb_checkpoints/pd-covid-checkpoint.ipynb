```python
import pandas as pd
from io import StringIO

df = pd.read_csv('WHO-COVID-19-global-data.csv',index_col='Date_reported', usecols=['Date_reported', 'Country', 'Cumulative_cases'])
print(df)
```

```python
df_brazil = df[df['Country']=='Brazil']
print(df_brazil)
```

```python
df_brazil = df[df['Country']=='Brazil']
print(df_brazil.shape)
df_italy = df[df['Country']=='Italy']
print(df_italy.shape)
df_USA = df[df['Country']=='United States of America']
print(df_USA.shape)
```

```python
df_brazil.to_csv('WHO-COVID-19-global-data-brazil.csv')
df_italy.to_csv('WHO-COVID-19-global-data-italy.csv')
df_USA.to_csv('WHO-COVID-19-global-data-USA.csv')
```

```python
import csv
import matplotlib.pyplot as plt
import matplotlib as mpl

x1 = []
y1 = []


with open('WHO-COVID-19-global-data-brazil.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots)
    for row in plots:
        x1.append(row[0])
        y1.append(int(row[2]))

        
plt.plot_date(x1,y1, xdate = True, linestyle="solid", marker="x", color="green")

mpl.rcParams["figure.autolayout"] = True
plt.xlabel('date')
plt.xticks(rotation=90)
plt.ylabel('cases')
plt.ylim([0,None])
plt.title('COVID- 19 CASES')
plt.subplots_adjust(left=0.08, right=0.97, bottom=0.25, top=0.92)
plt.grid(b=True, which='major', axis='both')
plt.tight_layout()
plt.savefig('covidcasesbrazil.png', dpi = 150)
plt.show()
```

```python
x2 = []
y2 = []

with open('WHO-COVID-19-global-data-italy.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots)
    for row in plots:
        x2.append(row[0])
        y2.append(int(row[2]))
        
plt.plot_date(x2,y2, xdate = True, linestyle="solid", marker="x", color="blue")

#mpl.rcParams["figure.figsize"] = 25, 10
mpl.rcParams["figure.autolayout"] = True
mpl.rcParams["axes.labelpad"] = 20
plt.xlabel('date')
plt.xticks(rotation=90)
plt.ylabel('cases')
plt.ylim([0,None])
plt.tick_params(axis='both',pad=5.0, labelsize='medium')
plt.title('COVID- 19 CASES')
plt.subplots_adjust(left=0.08, right=0.97, bottom=0.25, top=0.92)
plt.grid(b=True, which='major', axis='both')
plt.tight_layout()
plt.savefig('covidcasesitaly.png', dpi = 150)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import matplotlib as mpl

x1 = []
y1 = []


with open('WHO-COVID-19-global-data-brazil.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots)
    for row in plots:
        x1.append(row[0])
        y1.append(int(row[2]))
        
x2 = []
y2 = []

with open('WHO-COVID-19-global-data-italy.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots)
    for row in plots:
        x2.append(row[0])
        y2.append(int(row[2]))

x3 = []
y3 = []       
        
with open('WHO-COVID-19-global-data-USA.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots)
    for row in plots:
        x3.append(row[0])
        y3.append(int(row[2]))

plt.plot_date(x3,y3, xdate = True, linestyle="solid", marker="x", color="red", label='USA')  
plt.plot_date(x2,y2, xdate = True, linestyle="solid", marker="x", color="blue", label='Italy')        
plt.plot_date(x1,y1, xdate = True, linestyle="solid", marker="x", color="green", label='Brazil')

mpl.rcParams["figure.autolayout"] = True
mpl.rcParams["axes.labelpad"] = 20
mpl.rcParams["legend.loc"] = 'best'
#mpl.rcParams["figure.figsize"] = [25, 15]
plt.xlabel('date')
plt.xticks(rotation=90)
plt.ylabel('cases')
plt.ylim([0,None])
plt.tick_params(axis='both',pad=5.0, labelsize='medium')
plt.title('BRA X ITA X USA - COVID- 19 CASES')
plt.subplots_adjust(left=0.08, right=0.97, bottom=0.25, top=0.92)
plt.grid(b=True, which='major', axis='both')
plt.tight_layout()
plt.legend()
plt.show()
```
