### Modify dataset

```python
dataset['size'] = (dataset['bedrooms']* 20)
```

```python
dataset['size']
```

```python
# defining a function to categorize the size of bedrooms
def categoriza(s):
    if s >= 80:
        return 'Big'
    elif s >= 60:
        return 'Medium'
    elif s >= 40:
        return 'Small'
```

```python
# apply the fuction to size collumn and atribute to a new collumn
dataset['cat_size'] = dataset['size'].apply(categoriza)
dataset['cat_size']
```

```python
# count the distribution 
pd.value_counts(dataset['cat_size'])
```

```python
# option axis=1 define to drop a collumn
dataset.drop(['cat_size'], axis=1, inplace=True)
dataset.drop(['size'], axis=1, inplace=True)
```

```python
# drop lines by collumn conditions
dataset.drop(dataset[dataset.bedrooms==0].index ,inplace=True)
dataset.drop(dataset[dataset.bedrooms>30].index ,inplace=True)
```

```python
# missing values
dataset.isnull().sum()
```

```python
# remove all rows where there is at least one record missing in some attribute
# parameter inplace that the original data can be modified without creating a copy
dataset.dropna(inplace=True)
```

```python
# rows that are missing values in all columns
dataset.dropna(how='all', inplace=True)
```

```python
# fill in the average of the values of the floor columns the values null
dataset['floors'].fillna(dataset['floors'].mean(), inplace=True)
```

```python
# fill with 1 the null values ​​of the bedrooms column
dataset['bedrooms'].fillna(1, inplace=True)
```

```python
pd.value_counts(dataset['view'])
```
